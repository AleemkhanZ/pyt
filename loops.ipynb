```python
#reverse of a nunber 
num= int(input())
org= num
reverse = 0

while num>0:
    digit = num%10
    reverse = reverse*10+digit
    num = num//10

print("original number:",org)
print("Reverse number:" , reverse)
```

```python
#palindrome of a nunber 
num= abs(int(input()))
org= num
reverse = 0

while num>0:
    digit = num%10
    reverse = reverse*10+digit
    num = num//10

if org == reverse:
    print("Palindrome")
else:
    print("not palindrome")
```

```python
#Armstrong number 
num = int(input())
org = num
digits = len(str(num))
sum_of_powers = 0

while num>0:
    digit = num%10
    sum_of_powers += digit**digits
    num=num//10
    
if org == sum_of_powers:
    print("Armstrong number")
else:
    print("Not Armstrong number") 
```

```python
# fibonacci series
n = int(input())
a = 0 
b = 1

for i in range(n):
    print(a, end=" ")
    c=a+b
    a=b
    b=c
```

```python
# REVERSE THE NUMBER
n = int(input())
reverse = 0
Absnum = abs(n)
if n>=0:
    while n>0:
        digit = n%10
        reverse = reverse*10+digit
        n = n//10
    print(reverse)
else:
    while Absnum>0:
        digit = Absnum%10
        reverse = reverse*10+digit
        Absnum = Absnum//10
    print(reverse-2 * reverse)
```
