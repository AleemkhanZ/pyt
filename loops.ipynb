```python
fruits = ["apple" ,"mango", "cherry","onion","banana","guava"]
for i in fruits:

    if i == "onion":
        break
    print(i)
```

```python
fruits = ["apple" ,"mango", "cherry","onion","banana","guava"]
for i in fruits:
    if i == "onion":
        continue
    print(i)
```

```python
n = int(input())
if n<=1:
    print("Not a Prime Number")
else:
    for i in range(2,n):
        if n%i==0:
            print("not prime")
            break
    else:
        print(" Prime Number")
        
```
