CONDITITON STATEMENET 
- IF : Multiple Conditions 
- ELIF : Multiple Conditions
- ESLE IF : Single Conditions
- NESTED IF : Multiple Condition but with different concept

```python
marks = 85 
if marks >= 90:
    print("A+")
if  marks >= 75:
    print("A")
if marks >= 60:
    print("B")
if marks <60:
    print("Fail")
```

```python
marks = 85 
if marks >= 90:
    print("A+")
elif  marks >= 75:
    print("A")
elif marks >= 60:
    print("B")
elif marks <60:
    print("Fail")
else:
    print("Invalid marks")
```

```python
age = 90 
if age >= 18:
    print('You are eligible to vote')
else:
    print('you are not eligible to vote')
```
