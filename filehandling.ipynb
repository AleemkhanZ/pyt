```python
# Write mode
file = open( "data.txt" , "w" )
file.write("This is a new line a text.\n")
file.close()
```

```python
# Reading a file
file = open("data.txt" , "r") 
content = file.read()
print(content)
file.close()
```

```python
file = open( "data.txt" , "x" )
file.close()
```

```python
file = open( "data.txt" , "a" )
file.write("This is an appended line of text.\n")
file.close()
```
