```python
# Write mode
file = open( "data.txt" , "w" )
file.write("This is a new line a text.\n")
file.close()
```

```python
# Reading a file
file = open("data.txt" , "r") 
content = file.read()
print(content)
file.close()
```

```python
file = open("data.txt" , "x" )
file.close()
```

```python
#append 
file = open( "data.txt" , "a" )
file.write("This is an appended line of text.\n")
file.close()
```

```python
# shortcut for file.close() = with ...... as file:
with open("data.txt", "r") as file: # shortcut for close file 
    content = file.read()
    print(content)
 
```

```python
with open("data.txt", "r") as file:
    content = file.readline() # read first line
    print(content)
```

```python
file = open("data.txt", "r")
print(file.name) # name if the file
print(file.mode) # mode in which the file is opened
print(file.closed) # check if the file is closed or not 
file.read(5)
print(file.tell()) # current position of the file cursor
file.seek(0) # move the file cursor to the beinning of the file
```

#EXCEPTION HANDLING

```python
age= int(input("Enter your age:"))
if age>18:
    print("Valid")
else:
    print("Invalid")
```

```python
try:
    age= int(input("Enter your age:"))
    print(age)
except ValueError:
    print("Please enter a valid age")
    
```

```python
try:
    file = open('newfile,txt','x')
    print("File created successfully.")
except FileExistsError:
    print("File already exists")
finally:
    file.close()
```
