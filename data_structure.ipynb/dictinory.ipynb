```python
student = {
    "name" : "Rahul",
    "age" : 20,
    "course" : "BCA"
}
print(student)
```

```python
temp = {
    "lucknow" : [28,30,35],
    "chennai" : [25,28,30],
    "mumbai" : [40,45,38]
}
print(temp)
```

```python
info = dict(name = "Joy" , course="BCA", age=20)
print(info)
```

```python
#accessing values using keys

#1st method 
student['name']
student['age']
```

```python
#2nd method 
student.get['name']
student.get['age']
```

```python
print(student.keys())
print(student.values())
print(student.items())
```

```python
# adding a new pair
student['city'] = "Lko"
print(student)
```

```python
#deleting a pair 
del student['city']
print(student)
```

```python
student['course'] = "BBA"
print(student)
```

```python
for i in student:
    print(i)
```

```python
for i,v in student.items():
    print(f'{i} : {v}')
```

```python
company = {
    "emp001" : {
        "name" : 'Roy' , 
        "salary" : 40000 , 
        "Department" : "HR"
    },
    "emp002" : {
        "name" : 'Raj' , 
        "salary" : 60000 , 
        "Department" : "IT"
    }
}
print(company)
```

```python
company['emp003'] = {
    "name": "Gaurav",
    "salary" : 30000 , 
    "Department" : 'Sales'

}
print(company)
```

```python
company['emp002']["name"] = "Raj yadav" 
print(company)
    
```

```python
temp = {}
n = int(input("Enter the number of days: "))

for i in range(n):
    days = input("Enter day: ")
    temperature = int(input("Enter temperature: "))
    temp[days] = temperature

total = 0
for value in temp.values():
    total = total + value 
    average = total/len(temp.values())
print(average)
```
