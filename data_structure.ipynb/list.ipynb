**INDEXING AND SLICING**

```python
apps = ['google','swiggy','zomato','drive','flipkart','googlemap','amazon'] 
print(apps[1])
print(apps[5])
print(apps[-2])
print(apps[-5])
```

```python
print(apps[1:6])
print(apps[1:6:2])
print(apps[:6])
print(apps[5:])
print(apps[:])
```

**LIST METHOD**

LIST METHODS
- append  - add an element at the end of the list
- insert - add an element at any specfic position
- pop - deletes an element from the end and also using index value
- remove - deletes an element using the name of it 
-clear - clear whole list
- copy - make a copy of the list 
- sort - sort the element in specfic order
- extend - conbine two list

```python
apps.append('netflix')
print(apps)
```

```python
apps.insert(2,'Snapchat')
print(apps)
```

```python
apps.pop(2)
print(apps)
```

```python
apps = ['google','swiggy','zomato','drive','flipkart','googlemap'] 
apps2 = apps.copy()
print(apps)
```

```python
apps2.clear()
print(apps2)
```

```python
apps.sort()
print(apps)
```

```python
lst = [1,2,3,4,5]
apps.extend(lst)
print(apps)
```

```python
lst.extend(apps)
print(lst)
```
